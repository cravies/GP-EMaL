# Exploratory Data Analysis

```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
old = pd.read_csv("~/tree_stats_old.csv")
old.head()
```

```python
old.tail()
```

```python
new = pd.read_csv("~/tree_stats_new.csv")
new.tail()
```

```python
# let's drop bad columns
new = new.drop(['filename','algo'], axis=1)
old = old.drop(['filename','algo'], axis=1)
```

```python
# plotting parameters
BINS=10
```

```python
fig, axs = plt.subplots(2,3, figsize=(10, 10))  # create a 3x2 subplot
for i, col in enumerate(new.columns):  # loop over the columns
    ax = axs[0 if (i<3) else 1, i%3]  # get the corresponding axis
    ax.tick_params(labelrotation=45)
    ax.xaxis.set_major_locator(plt.MaxNLocator(BINS)) # set number of bins
    ax.axvline(np.median(new[col]), color='red', linestyle='dashed', linewidth=3) # median line
    ax.hist(new[col], bins=BINS)  # create a histogram of the column data
    ax.set_title(f"GPEMaL {col} count")  # set the title of the subplot
    ax.set_xlabel('Value')  # set the x-label
    ax.set_ylabel('Frequency')  # set the y-label

plt.tight_layout()  # adjust the spacing between subplots
plt.show()  # show the plot
```
